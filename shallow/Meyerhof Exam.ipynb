### Meyerhof (1963) Bearing Capacity Equation

<center> $ q_{u} = c^{'} N_{c} F_{cs} F_{cd} F_{ci} + q N_{q} F_{qs} F_{qd} F_{qi} + 0.5 \gamma B N_{r} F_{ \gamma s} F_{ \gamma d} F_{\gamma i}  $ </center> 

### Shape Factor

<center> $ F_{cs} = 1 + \frac{B}{L} \frac{N_{q}}{N_{c}} $ </center> <br>
<center> $ F_{qs} = 1 + \frac{B}{L} \text{tan} \phi^{'} $ </center> <br>
<center> $ F_{ys} = 1 + 0.4\frac{B}{L} $ </center>

```python
def meyerhofShapeFactor(b,l,nq,nc,angle):
    fcs = 1 + (b/l)*(nq/nc)
    fqs = 1 + (b/l)*math.tan(math.radians(angle)) 
    fys = 1 - (0.4*(b/l))
        
    return {
        "fcs" : fcs, 
        "fqs" : fqs,
        "fys" : fys
            
}
shape_factor = meyerhofShapeFactor(data["B"],data["B"],nq,nc,angle)
fcs = shape_factor["fcs"]
fqs = shape_factor["fqs"]
fys = shape_factor["fys"]
print("fcs = " + str(shape_factor["fcs"]))
print("fqs = " + str(shape_factor["fqs"]))
print("fys = " + str(shape_factor["fys"]))
```

### Depth Factor

$ \frac{D_{f}}{B} < 1 $ <br>

For $ \phi $ = 0: 

<center> $ F_{cd} = 1 + 0.4 \text{tan}^{-1} \left( \frac{d_{f}}{B} \right) $  </center> <br>
<center> $ F_{qd} = 1 $  </center> <br>
<center> $ F_{yd} = 1 $  </center> 

For $ \phi $ > 0: 

<center> $ F_{cd} = F_{qd} - \frac{ 1 - F_{qd} }{ N_{c} \text{tan} \phi ' } $  </center> <br>
<center> $ F_{qd} = 1 + 2 \text{tan}\phi^{'} \left(1 - \text{sin} \phi' \right)^{2} $  </center> <br>
<center> $ F_{yd} = 1 $  </center> 


$ \frac{D_{f}}{B} > 1 $ <br>

For $ \phi $ = 0: 

<center> $ F_{cd} = 1 + 0.4 \text{tan}^{-1} \left( \frac{B}{L} \right) $  </center> <br>
<center> $ F_{qd} = 1 $  </center> <br>
<center> $ F_{yd} = 1 $  </center> 

For $ \phi $ > 0: 

<center> $ F_{cd} = F_{qd} - \frac{ 1 - F_{qd} }{ N_{c} \text{tan} \phi ' } $  </center> <br>
<center> $ F_{qd} = 1 + 2 \text{tan}\phi^{'} \left(1 - \text{sin} \phi' \right)^{2} \text{tan}^{-1} \frac{D_{f}}{B} $  </center> <br>
<center> $ F_{yd} = 1 $  </center> 



```python
def meyerhofDepthFactor(df,b,angle,nc):

    dfoverB = df/b
#     print(dfoverB)
    rad_angle = math.radians(angle)
    if dfoverB <= 1:
        if angle == 0 :
            fcd = 1 + (0.4*(dfoverB))
            fqd = 1
            fyd = 1
        else:
            fqd =  1 + ((2*math.tan(rad_angle))*math.pow(1 - math.sin(rad_angle),2)*dfoverB)
            fcd = fqd - ((1 - fqd)/(nc*math.tan(rad_angle)))
            fyd  = 1
    else :
        if angle == 0:
            fqd_init = 0.4*math.atan(dfoverB) ##radians
            fqd =  1 + (0.4*fqd_init)
            fyd  = 1
            fcd = 1 
         
        else:
            fqd_init = math.atan(dfoverB) ##radians
            fqd =  1 + 2*math.atan(rad_angle)*math.pow(1 - math.sin(rad_angle),2)*fqd_init
            fcd = 1 + (0.4*(dfoverB))
            fyd  = 1
     
    return {
        "fcd" : fcd,
        "fqd" : fqd,
        "fyd" : fyd,               
    }
depth_factor = meyerhofDepthFactor(data["Df"],data["B"],angle,nc)
# depth_factor = meyerhofDepthFactor(1.22,1,34,nc)
fcd = depth_factor["fcd"]
fqd = depth_factor["fqd"]
fyd = depth_factor["fyd"]
print("fcd = " + str(fcd))
print("fqd = " + str(fqd))
print("fyd = " + str(fyd))
```
